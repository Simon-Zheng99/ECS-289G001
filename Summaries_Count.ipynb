```python
import matplotlib.pyplot as plt
```

```python
# Insightful summaries Phase1
x1 = [0, 1, 2, 3, 4, 5, 6]
y1 = [0, 0, 0, 0, 2.5, 31, 100]
```

```python
plt.figure(figsize=(8,5))
plt.plot(x1, y1, marker='o', linestyle='-', color='r')

plt.xlabel('Prediction Head')
plt.ylabel('Number of Insightful Summaries')
plt.title('Insightgul Summaries for Phase 1')

plt.grid(True)
plt.savefig("summaries_1.jpg")
plt.show()
```

```python
# Insightful summaries Phase2
x2 = [0, 1, 2, 3, 4, 5, 6]
y2 = [0, 7.5, 9, 15.5, 38.5, 15.5, 61.5]
```

```python
plt.figure(figsize=(8,5))
plt.plot(x2, y2, marker='o', linestyle='-', color='b')

plt.xlabel('Prediction Head')
plt.ylabel('Number of Insightful Summaries')
plt.title('Insightgul Summaries for Phase 2')

plt.ylim(0, 100)

plt.grid(True)
plt.savefig("summaries_2.jpg")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.plot(x1, y1, marker='o', linestyle='-', color='r', label='Phase 1')
plt.plot(x2, y2, marker='o', linestyle='-', color='b', label='Phase 2')

plt.xlabel('Prediction Head')
plt.ylabel('Number of Insightful Summaries')
plt.title('Comparison of Insightful Summaries (Phase 1 & 2)')

plt.ylim(0, 100)

plt.grid(True)
plt.legend()

plt.savefig("combined_summaries.jpg")
plt.show()
```
